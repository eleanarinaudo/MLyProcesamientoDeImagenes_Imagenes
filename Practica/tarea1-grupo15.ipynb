## Ejercicio 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Construir las siguientes banderas (tamaño 256x512). Bandera a elección en tonalidades de grises
flag = np.zeros((256, 512))

flag[:, 170:340] = 1
print("La dimensión de la bandera es: ", flag.shape)
plt.imshow(flag, cmap="gray")
plt.show()
```

```python
decreciente = np.linspace(0, 1, 512)

for i in range(256):
    flag[i, :] = decreciente  # sobreescribo cada fila: fila i, todas las columnas

flag[:256, :] = decreciente
flag[:, 170:340] = 1

plt.imshow(flag, cmap="gray")
plt.show()
```

```python
# Bandera a elección en colores
flag_rgb = np.zeros((256, 512, 3), "uint8")
flag_rgb[0:256, :170, 2] = 255
flag_rgb[0:256, 170:340] = 255
flag_rgb[0:256, 340:, 0] = 255

plt.imshow(flag_rgb)
plt.show()
```


- Elegir alguna de las transformaciones geométricas vistas en clase (1 o mas) y aplicar a la imagen del Ejercicio 1 mostrando imagen original e imagen transformada en subplots. Hacer lo mismo pero con  la imagen  del Ej2.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# from scipy import signal
from scipy.ndimage import rotate
```

```python
from google.colab.patches import cv2_imshow
```

# Bandera


```python
# Recorte
flag_r = flag[60:236, 74:376]  # Un recorte de interés
print("Tamaño original:", flag.shape, "\nTamaño recorte:", flag_r.shape)
plt.figure(figsize=(10, 8))
plt.subplot(121), plt.imshow(flag, cmap="plasma"), plt.title("Bandera Completa"),
plt.subplot(122), plt.imshow(flag_r, cmap="plasma"), plt.title("Bandera recortada ")
plt.show()
```

```python
# TRASLACION

IGcirc = np.fft.fftshift(flag)
# NOTAR fftshift SOLO TRASLADA (CIRCULARMENTE)

plt.figure(figsize=(12, 100))
plt.subplot(121), plt.imshow(flag, cmap="magma"), plt.title("Bandera")
plt.subplot(122), plt.imshow(IGcirc, cmap="magma"), plt.title(
    "Imagen  trasladada circularmente"
)
plt.show()
```

```python
# ROTACIÓN

height, width = flag.shape

R = cv2.getRotationMatrix2D(
    (width / 2, height / 2), 45, 0.5
)  # Los parámetros width/2,height/2, indican que la rotación se haga desde el centro
flagR45 = cv2.warpAffine(flag, R, (width, height))
flagR45dim = rotate(
    flag, 45
)  # Se mantienen las mismas dimensiones de la imagen original. La función rotate es de la librería scipy Y NO PONEMOS FALSE se acomoda con el ancho y altura original

print("dimensiones original", flagR45.shape)
print("dimensiones rotada manteniendo dimensiones de la original", flagR45dim.shape)

plt.figure(figsize=(15, 12))
plt.subplot(131), plt.imshow(flag, cmap="inferno"), plt.title("Bandera")
plt.subplot(132), plt.imshow(flagR45, cmap="inferno"), plt.title(
    "Bandera original rotada 45 escalada a la mitad"
)
plt.subplot(133), plt.imshow(flagR45dim, cmap="inferno"), plt.title(
    "Bandera original rotada 45 mantiendo dimensiones"
)
plt.show()
```

# Imagen 

```python
# Recorte
bambi = cv2.imread("pics/bambi.jpg", 0)
print(bambi.shape, type(bambi))

bambi_r = bambi[74:376, 90:200]
print("Tamaño original:", bambi.shape, "\nTamaño recorte:", bambi_r.shape)
plt.figure(figsize=(20, 8))
plt.subplot(121), plt.imshow(bambi), plt.title("Imagen Bambi Completa"),
plt.subplot(122), plt.imshow(bambi_r), plt.title("Imagen Bambi Recortada ")
plt.show()
```

```python
# TRASLACION

IGcirc = np.fft.fftshift(bambi)
# NOTAR fftshift SOLO TRASLADA (CIRCULARMENTE)

plt.figure(figsize=(12, 100))
plt.subplot(121), plt.imshow(bambi), plt.title("Bambi Original")
plt.subplot(122), plt.imshow(IGcirc), plt.title("Imagen  trasladada circularmente")
plt.show()
```

```python
# ROTACIÓN

height, width = bambi.shape

R = cv2.getRotationMatrix2D(
    (width / 2, height / 2), 90, 0.5
)  # Los parámetros width/2,height/2, indican que la rotación se haga desde el centro
bambiR90 = cv2.warpAffine(bambi, R, (width, height))
bambiR90dim = rotate(
    bambi, 90
)  # Se mantienen las mismas dimensiones de la imagen original. La función rotate es de la librería scipy Y NO PONEMOS FALSE se acomoda con el ancho y altura original

print("dimensiones original", bambiR90.shape)
print("dimensiones rotada manteniendo dimensiones de la original", bambiR90dim.shape)

plt.figure(figsize=(15, 12))
plt.subplot(131), plt.imshow(bambi, cmap="inferno"), plt.title("Bambi")
plt.subplot(132), plt.imshow(bambiR90, cmap="inferno"), plt.title(
    "Bambi rotada 90 escalada a la mitad"
)
plt.subplot(133), plt.imshow(bambiR90dim, cmap="inferno"), plt.title(
    "Bambi rotada 90 mantiendo dimensiones"
)
plt.show()
```
